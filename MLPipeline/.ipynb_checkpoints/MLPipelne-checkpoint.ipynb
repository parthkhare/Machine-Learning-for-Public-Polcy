```python
def treat_missing(df):
    '''
    is 0 a missing value ? Na/NaN
    droppig categorical missing  ? (replace median/or drop row ?)

    Shows missing value frequency in data, fill NA values with respective feature mean
    Inputs: pandas DataFrame(df)
    Output: DataFrame without missing values(df_nm)
    '''
    # Find Missing Values
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    print('Missing data summary')
    display(missing_data.head(20))
    
    # Replace the NaN with mean value
    df_nm = df.apply(lambda x: x.fillna(x.mean()),axis=0)
    
    return df_nm

# Call
# df_clean = treat_missing(df)
```

## discretize a continuous variable

```python
def cont_discr(df, cont_var, bins=5):
    '''
    Discretize a continuous variable in the data set, by quantile cut or regular otherwise.
    Inputs: pandas DataFrame(df)
            features to examine interactions of(cont_var)
            number of bins(bins, default=5)
    Output: Transformed DataFrame(df_t)
    '''
    # Create copy of the data to augment variables into
    df_tr = df.copy(deep=True)
    for feature in cont_var:
        bin_feature = 'cat_'+ feature
        df_tr[bin_feature] = pd.qcut(df[feature], bins, labels=list(range(bins)))
        df_tr[bin_feature] = df_tr[bin_feature].astype('str')  

    return df_tr 
```
